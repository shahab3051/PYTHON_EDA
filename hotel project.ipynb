# Data analysis and visualization 

```python
canceled_per=df['is_canceled'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(6,3))
plt.title("Reservation Status Count")
plt.bar(['Not canceled',"Canceled"],df['is_canceled'].value_counts(),width=0.8,color="Skyblue")
plt.show()
```

```python
plt.figure(figsize=(10,6))
lab=sns.countplot(data=df,x='hotel',hue='is_canceled',palette="mako")
plt.bar_label(lab.containers[0])
plt.bar_label(lab.containers[1])
plt.legend(["not Canceled","Canceled"])
```

```python
df.columns
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
plt.figure(figsize=(15,8))
plt.title("Average daily rate in city and resort hotel",fontsize=20)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=15)
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(10,5))
lb=sns.countplot(x='month',hue='is_canceled',data=df,palette='BuPu')
plt.bar_label(lb.containers[0])
plt.bar_label(lb.containers[1])
plt.title("Reservation Status per month")
plt.legend(["not_canceled","canceled"])
```

```python
canceled_data=df[df["is_canceled"]==1]
top_10_country =canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.title("Top 10 countries with reservation canceled")
```

```python
canceled_data=df[df["is_canceled"]==0]
top_10_country =canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.title("Top 10 countries with successful reservations")
```

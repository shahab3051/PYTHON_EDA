```python
plt.figure(figsize=(10,5))
sns.lineplot(x='pizza_size',y="total_price",data=df,marker='o')
plt.title("Total_sales")
plt.xlabel("pizza_size")
plt.ylabel("total_price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='pizza_category',y='unit_price',data=df,palette='magma')
plt.title("Pizza_Category")
plt.xlabel('pizza_category')
plt.ylabel('unit_price')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='pizza_name',y='unit_price',data=df,palette='flare')
plt.title("Pizza_Category")
plt.xlabel('pizza_category')
plt.ylabel('unit_price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='pizza_size',y="quantity",data=df,palette='flare')
plt.title("Total_Quantity")
plt.xlabel("pizza_size")
plt.ylabel("Quantity of items")
plt.show()
```
